```python
import matplotlib.pyplot as pl
import numpy as np

# -pi부터 pi까지 포함하여 균일하게 분포된 256개의 점 생성
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)

# 이것들은 내장된 Python 수학의 벡터화된 버전입니다.
# 각 x에 대해 코사인 계산
y = np.cos(x)

# 각 x에 대해 사인 계산
y1 = np.sin(x)

# 코사인과 사인 모두 그림
pl.plot(x, y)
pl.plot(x, y1)

pl.show()
```

```python
from pylab import *
import numpy as np

# -pi부터 pi까지 포함하여 균일하게 분포된 256개의 점 생성
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)

# 이것들은 내장된 Python 수학의 벡터화된 버전입니다.
# 각 x에 대해 코사인 계산
y = np.cos(x)

# 각 x에 대해 사인 계산
y1 = np.sin(x)

# 코사인 그리기
plot(x, y)

# 사인 그리기
plot(x, y1)

# 그래프 제목 정의
title("Functions $\sin$ and $\cos$")

# x 축 한계 설정
xlim(-3.0, 3.0)
# y 축 한계 설정
ylim(-1.0, 1.0)

# 특정 값에서 눈금 포맷팅
xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
yticks([-1, 0, +1],
          [r'$-1$', r'$0$', r'$+1$'])

show()
```
