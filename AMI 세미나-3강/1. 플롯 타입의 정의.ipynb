```python
from matplotlib.pyplot import *

# 간단한 데이터
x = [1,2,3,4]
y = [5,4,3,2]

# 새로운 피규어 생성
figure()

# subplot을 2x3 그리드로 나누고 1번 선택
subplot(231)
plot(x, y)

# 2번을 선택
subplot(232)
bar(x, y)

# 수평막대그래프
subplot(233)
barh(x, y)

# 누적 막대 그래프 생성
subplot(234)
bar(x, y)

# 누적 막대 그래프를 위해 추가 데이터 필요
y1 = [7,8,5,3]
bar(x, y1, bottom=y, color = 'r')

# 박스그래프
subplot(235)
boxplot(x)

#산포도
subplot(236)
scatter(x,y)

show()
```
