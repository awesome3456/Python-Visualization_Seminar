```python
from openpyxl import load_workbook
from pprint import pprint
from datetime import datetime

# 'ch02-xlsxdata.xlsx' 파일을 열기
file = 'ch02-xlsxdata.xlsx'
wb = load_workbook(file)
ws = wb['Sheet1']

dataset = []

# openpyxl을 사용하여 데이터 읽기
for row in ws.iter_rows(values_only=True):
    dataset.append(list(row))

pprint(dataset)

# (2, 1) 위치의 셀 정보 출력 (엑셀 셀 인덱스는 1부터 시작)
cell = ws.cell(2, 1)
print(cell)           # 셀 정보 출력
print(cell.value)     # 셀 값 출력

# 셀이 날짜 형식인지 확인 후, 날짜를 출력
if cell.data_type == 'd':
    date_value = cell.value
    print(date_value)  # 날짜 출력
```
